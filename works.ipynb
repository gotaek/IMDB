```python
import csv
import pymysql
```

```python
f = open('title.basics.tsv', 'r', encoding='utf=8')
rd = csv.reader(f, delimiter='\t')

conn=pymysql.connect(host='localhost',user='gotaek',password='gotaek',db='IMDB')
curs=conn.cursor(pymysql.cursors.DictCursor)

sql="insert ignore into works (workId,titleType,primaryTitle,originalTitle,isAdult,startYear,endYear,runtimeMinutes) values(%s,%s,%s,%s,%s,%s,%s,%s)"

next(rd)

for line in rd:
    isAdult=line[4]
    startYear=line[5]
    endYear=line[6]
    runtimeMinutes=line[7]
    for i in range(4,8):
        if line[4]!="0":
            isAdult=1
        if line[5]=="\\N":
            startYear=0
        if line[6]=="\\N":
            endYear=0
        if line[7]=="\\N":
            runtimeMinutes=0
    data=(line[0],line[1],line[2],line[3],isAdult,startYear,endYear,runtimeMinutes)
    print(data)
    
    curs.execute(sql,data)
    conn.commit()
f.close()
```
