```python
import csv
import pymysql
```

```python
f = open('title.akas.tsv', 'r', encoding='utf=8')
rd = csv.reader(f, delimiter='\t')

conn=pymysql.connect(host='localhost',user='gotaek',password='gotaek',db='IMDB')
curs=conn.cursor(pymysql.cursors.DictCursor)

sql="insert ignore into akas(workId,ordering,title,region,lang,isOriginalTitle)values(%s,%s,%s,%s,%s,%s)"

next(rd)
for line in rd:
    data=(line[0],line[1],line[2],line[3],line[4],line[7])
    print(data)
    curs.execute(sql,data)
    conn.commit()
   
    

curs.close()
conn.close()
f.close()
```
