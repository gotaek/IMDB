```python
import csv
import pymysql
```

```python
f = open('title.akas.tsv', 'r', encoding='utf=8')
rd = csv.reader(f, delimiter='\t')

conn=pymysql.connect(host='localhost',user='gotaek',password='gotaek',db='IMDB')
curs=conn.cursor(pymysql.cursors.DictCursor)

sql="insert ignore into akas_types(workId,ordering,akasType)values(%s,%s,%s)"

next(rd)

for line in rd:
    for i in line[5].split(','):
        data=(line[0],line[1],i)
        print(data)
    curs.execute(sql,data)
    conn.commit()
   
    

curs.close()
conn.close()
f.close()
```
